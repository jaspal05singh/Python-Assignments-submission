```python

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df
```

```python
#1. Some values in the the FlightNumber column are missing. These numbers are
#meant to increase by 10 with each row so 10055 and 10075 need to be put in
#place. Fill in these missing numbers and make the column an integer column
#(instead of a float column).
df["FlightNumber"].fillna(0, inplace=True)
```

```python
df["FlightNumber"]= df["FlightNumber"].astype(int)
```

```python
l=[]

for i in range(len(df["FlightNumber"])):
    if(df["FlightNumber"][i]>0):
        l.append(df["FlightNumber"][i])
    else:
        l.append(l[i-1]+10)
df["FlightNumber"]= l
```

```python
#2. The From_To column would be better as two separate columns! Split each
#string on the underscore delimiter _ to give a new temporary DataFrame with
#the correct values. Assign the correct column names to this temporary
#DataFrame.
newTempDf= pd.DataFrame(columns=["From", "To"])
```

```python
from_loc=[]
to_loc=[]
for i in df["From_To"]:
    places= i.split("_")
    from_loc.append(places[0])
    to_loc.append(places[1])
print(from_loc)
print(to_loc)
```

```python
newTempDf["From"]= from_loc
newTempDf["To"]= to_loc
newTempDf
```

```python
#3. Notice how the capitalisation of the city names is all mixed up in this
#temporary DataFrame. Standardise the strings so that only the first letter is
#uppercase (e.g. "londON" should become "London".)
def stringFormation(s):
    s1=s[0].upper()
    s2=s[1:len(s)+1].lower()
    return (s1+s2)
```

```python
newTempDf["From"]= newTempDf["From"].apply(stringFormation)
newTempDf["To"]= newTempDf["To"].apply(stringFormation)
newTempDf
```

```python
#4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.
df.drop("From_To", axis=1, inplace=True)
new_df= pd.concat([df, newTempDf],axis=1)
new_df
```

```python
#5. In the RecentDelays column, the values have been entered into the DataFrame as a list. We would like each first value in 
# its own column, each second value in its own column, and so on. If there isn't an Nth value, the value should be NaN.
# Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted 
# RecentDelays column in df with delays.
delays= list(df["RecentDelays"].values)
delay_df=pd.DataFrame(delays, columns=["delay_1","delay_2","delay_3"])

new_df.drop("RecentDelays", axis=1, inplace=True)

final_df=pd.concat([new_df, delay_df], axis= 1)
final_df
```
